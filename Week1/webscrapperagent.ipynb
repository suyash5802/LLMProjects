```python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def message_for(website):
    return [
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt_prefix+website}
    ]
message_for(ed)
```

```python
def summarize(url):
    website=fetch_website_contents(url)
    response=openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=message_for(url)

    )
    return response.choices[0].message.content
summarize("https://cnn.com")
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
display_summary("https://cnn.com")
```

```python
system_prompt = """You are an expert email assistant.
Your job is to read an email body and suggest a short,
professional subject line for it.
Keep it under 10 words. Return ONLY the subject line, nothing else."""

def suggest_subject_line(email_body):
    """Takes email body text and returns a suggested subject line"""
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user",   "content": f"Please suggest a subject line for this email:\n\n{email_body}"}
    ]
    response = openai.chat.completions.create(
        model    = "gpt-4o-mini",
        messages = messages
    )
    return response.choices[0].message.content.strip()
```

```python
# Test email 1 — meeting request
email_1 = """
Hi John,
I wanted to check if you are free next Tuesday at 3pm for a quick 
30 minute catch up to discuss the Q3 project progress and align 
on next steps before the deadline.
Let me know what works for you.
Thanks,
Sarah
"""

# Test email 2 — invoice
email_2 = """
Dear Team,
Please find attached the invoice for the web development services 
provided in October 2024. The total amount due is $2,500 and 
payment is requested within 30 days.
Please confirm receipt.
Regards,
Alex
"""

# Test email 3 — job application
email_3 = """
Dear Hiring Manager,
I am writing to apply for the Machine Learning Engineer position 
at your company. I have 3 years of experience in deep learning,
NLP and computer vision. I have attached my resume for your review.
I would love to discuss how I can contribute to your team.
Best regards,
Suyash
"""

# Run it
print("Email 1 subject:", suggest_subject_line(email_1))
print("Email 2 subject:", suggest_subject_line(email_2))
print("Email 3 subject:", suggest_subject_line(email_3))
```
